```python
import os
import pandas as pd
```

# **1. Leitura dos dados**

```python
estados_path = "/home/usuario/Documents/Awari_EngDados/awari-engenharia-de-dados-docker/exercicios/municipios-estados/csv/estados.csv"
cidades_path = "/home/usuario/Documents/Awari_EngDados/awari-engenharia-de-dados-docker/exercicios/municipios-estados/csv/municipios.csv"
```

```python
df_estados = pd.read_csv(estados_path)
df_cidades = pd.read_csv(cidades_path)
```

# **2. Criando pastas para cada UF**

```python
unique_states = df_estados['uf'].unique()

base_folder = 'estados'

os.makedirs(base_folder, exist_ok=True)

for state_acronym in unique_states:
    state_folder = os.path.join(base_folder, state_acronym)
    os.makedirs(state_folder, exist_ok=True)
    print(f"Criando pasta para o estado: {state_acronym}")
```

# **2. Organizando cidades por estado num único CSV para cada UF**

```python
display(df_estados.head(), df_cidades.head())
```

```python
df_cidades_estados = pd.merge(df_cidades, df_estados[["codigo_uf", "uf"]], on = ["codigo_uf"], how = 'left')
df_cidades_estados
```

```python
def create_state_csv(state_uf, cities_states_df):
    state_folder = os.path.join(f'{base_folder}', state_acronym)
    state_csv_path = os.path.join(state_folder, 'cidades.csv')
    
    state_cities_df = cities_states_df[cities_states_df['uf'] == state_uf]
    state_cities_df.to_csv(state_csv_path, index=False)
    print(f"Criado CSV para cidades o estado {state_uf}: {state_csv_path}")

for state_uf in unique_states:
    create_state_csv(state_acronym, df_cidades_estados)
```
